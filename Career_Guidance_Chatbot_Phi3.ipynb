# Career Guidance Chatbot

## Initialising the Pre-trained model Phi-3 by microsoft

In colab please switch to GPU (T4), else you might get errors.
Downloading the model may take few minutes (max 5 mins)

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

```

## Just a simple function to decorate the response

```python
def format_response(response):
  r = response.split(' ')
  count = 0
  for word in r:
    print(word, end= ' ')
    count += 1
    if count == 10:
      print(" ")
      count = 0
```

## Query and Answer part, run the below code or cell to give input which is your general career query and the get the response.

```python
# below is the input code block
prompt = input('Enter your career query. Feel free! : ')

prompt = 'You are an AI career counseller, so help the student with their career by giving valuable response in kind words as if you are a priest in a single paragraph. Student Query:  ' + prompt
messages = [
    {"role": "system", "content": prompt},
]


# parameters for the model to generate response
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.2,
    "do_sample": False,
}


# below is the code to generate the response and print it
output = pipe(messages, **generation_args)
response = output[0]['generated_text']
print(" ")
print("AI Counceller: ")
format_response(response)

```
