```python
from diffusers import DiffusionPipeline
import matplotlib.pyplot as plt

pipe = DiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda")

prompt = input("Enter the text to generate image: ")
print("Generating image....")
image = pipe(prompt).images[0]

image.save("/content/image.png")
plt.imshow(image)
plt.axis("off")
plt.show()

print("\n--- Rubric Evaluation ---")
criteria = {
    "Relevance to Prompt": 0,
    "Visual Quality": 0,
    "Creativity": 0,
    "Clarity/Coherence": 0
}

total_score = 0

for key in criteria:
    while True:
        try:
            score = int(input(f"Rate '{key}' (1-10): "))
            if 1 <= score <= 10:
                criteria[key] = score
                total_score += score
                break
            else:
                print("Please enter a number between 1 and 10.")
        except ValueError:
            print("Please enter a valid integer.")

print("\n--- Evaluation Results ---")
for key, value in criteria.items():
    print(f"{key}: {value}/10")
print(f"Total Score: {total_score}/40")
```
